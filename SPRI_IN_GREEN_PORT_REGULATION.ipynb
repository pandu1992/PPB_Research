```python
# Colab-ready notebook: Scrape port documents/pages and build SPRI per country/port
# Purpose: heuristic scrape to produce a Smart Port Readiness Index (SPRI)
# for Autonomous Ship Operations (2018–2024).

!pip install --quiet requests beautifulsoup4 pdfplumber pandas matplotlib scikit-learn tqdm pdfminer.six pdf2image pytesseract

import requests
from bs4 import BeautifulSoup
import pdfplumber
import io, re, json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
from pdfminer.high_level import extract_text as pdfminer_extract
from pdf2image import convert_from_bytes
import pytesseract

# ----------------------------
# Curated sources for ASEAN-5
# ----------------------------
sources = {
   'Singapore': [
    "https://www.mpa.gov.sg/maritime-singapore/sustainability/maritime-singapore-green-initiative",
    "https://www.mpa.gov.sg/web/portal/home/mpa-maritime-singapore/green-initiatives/green-port-programme",
    "https://www.mpa.gov.sg/maritime-singapore/sustainability/maritime-singapore-decarbonisation-blueprint",
    "https://www.mpa.gov.sg/docs/mpalibraries/mpa-documents-files/sustainability-office/mpa-decarb-blueprint-2050a.pdf",
    #"https://www.mot.gov.sg/what-we-do/green-transport/maritime-environment-responsibility/",
    "https://www.unescap.org/sites/default/d8files/event-documents/Sustainable_and_resilient_Pacific_Ports_Feb_2022_Final.pdf"
],
    'Malaysia': [
    "https://www.pka.gov.my/phocadownload/policy/PORT%20KLANG%20GREEN%20POLICY.pdf",
    "https://www.mot.gov.my/en/maritime",
    "https://www.mot.gov.my/en/Documents/Port%20Authorities%20Act%201963%20(act%20488)_2.pdf",
    "https://www.mot.gov.my/en/Documents/Ports%20(Privatization)%20Act%201990%20(act%20422).pdf",
    #"https://www.investmalaysia.gov.my/media/m3eamxey/thirteenth-malaysia-plan-2026-2030.pdf",
    "https://www.unescap.org/sites/default/d8files/event-documents/TS2_Country%20presentation_Malaysia.pdf"
],

    'Philippines': [
    "https://www.ppa.com.ph/content/Supporting%20Documents/Item%201.c_PPA%20AO%20No.%2005-2018%20%20The%20Port%20Environmental%20Policy%20%28PEP%29_0.pdf",
    "https://www.ppa.com.ph/",
    #"https://marina.gov.ph/wp-content/uploads/2024/04/MIDP-2028-as-of-29-April-2024.pdf",
    "https://www.denr.gov.ph/",
    "https://denr.gov.ph/news-events/green-shipping-coalition-roundtable-discussion/",
    "https://www.unescap.org/sites/default/d8files/event-documents/TS2_Country%20presentation_Philippines.pdf"
],
    'Thailand': [
    "https://pat.marketing/en/news/green-port/",
    "https://www.port.co.th/",
    "https://www.mot.go.th/",
    #"https://www.onep.go.th/",
    "https://md.go.th/wp-content/uploads/2021/03/full-translated_Thailands-Overall-Maritime-Strategy-as-of-21-01-2023.pdf",
    "https://www.unescap.org/sites/default/d8files/event-documents/18-Thailand.pdf"
],
    'Indonesia': [
    "https://peraturan.bpk.go.id/Details/103722/permenhub-no-29-tahun-2014",
    "https://peraturan.go.id/",
    "https://maritim.go.id/",
    #"https://jdih.esdm.go.id/",
    "https://perpustakaan.bappenas.go.id/e-library/file_upload/koleksi/migrasi-data-publikasi/file/Unit_Kerja/Dir%20Industri%2C%20Ekonomi%20dan%20Kreatif/Dummy_Indonesia%20Blue%20Economy%20Roadmap_Ebook.pdf",
    "https://www.unescap.org/sites/default/files/6.Indonesia.pdf"
],
   'Regional_Global_Reference': [
    "https://asean.org/",
    "https://asean.org/wp-content/uploads/2023/08/AMO-1.pdf",
    "https://www.unescap.org/",
    "https://www.imo.org/en/OurWork/Environment/Pages/IMO-Strategy-on-reduction-of-GHG-emissions-from-ships.aspx",
    "https://unctad.org/system/files/official-document/rmt2023_en.pdf",
    "https://www.oecd.org/environment/"
]
}

# ----------------------------
# ASEAN-5 Pillars (multilingual regex)
# ----------------------------
pillars = {
    'technological_infrastructure': [
        r'automated cargo', r'automated crane', r'shipboard crane', r'AGV\b',
        r'automated truck', r'automated handling', r'yard automation',
        r'automated stack', r'robotic', r'autonomous crane', r'smart yard',
        r'automated lashing', r'automated securing', r'auto[- ]lashing',
        r'otomatisasi kargo', r'crane otomatis', r'kendaraan otomatis',
        r'kendaraan berpandu otomatis', r'penanganan otomatis', r'yard pintar', r'robotik',
        r'awtomatikong kargamento', r'awtomatikong kreyn', r'robotiko',
        r'ระบบอัตโนมัติ', r'ปั้นจั่นอัตโนมัติ', r'ยานพาหนะอัตโนมัติ', r'หุ่นยนต์', r'ลานอัจฉริยะ'
    ],
    'advanced_autonomous_technologies': [
        r'Autonomous Surface Vessel', r'ASV\b', r'autonomous vessel', r'remote pilot',
        r'V2X', r'VTS', r'ASV berthing', r'autonomous berthing',
        r'autonomous navigation', r'collision avoidance system',
        r'kapal otonom', r'navigasi otonom', r'berthing otomatis', r'pengendali jarak jauh',
        r'sasakyang-dagat na awtonomo', r'nabigasyon na awtonomo',
        r'เรืออัตโนมัติ', r'การนำร่องระยะไกล', r'ระบบหลีกเลี่ยงการชน'
    ],
    'interoperability_communication': [
        r'API\b', r'interoperab', r'integrat(ed|ion) system', r'port community system',
        r'PCS\b', r'IEC 61968', r'IEC 61162', r'open standard',
        r'DATA SHARING', r'port-ship interface', r'protocol',
        r'semantic interoperability',
        r'sistem terintegrasi', r'sistem komunitas pelabuhan',
        r'sharing data', r'protokol',
        r'nakapagsasama', r'sistemang pinagsama', r'pagbabahagi ng datos',
        r'การเชื่อมต่อระบบ', r'มาตรฐานเปิด', r'การแบ่งปันข้อมูล', r'โปรโตคอล'
    ],
    'onshore_operation_centers': [
        r'Onshore Operation Centre', r'OOC\b', r'Onshore Control', r'remote control centre',
        r'remote operator', r'situational awareness', r'remote piloting',
        r'remote monitoring', r'control centre',
        r'pusat kendali darat', r'pusat operasi darat', r'pusat pengendalian',
        r'operator jarak jauh',
        r'sentro ng operasyon sa lupa', r'kontrol na malayuan',
        r'ศูนย์ควบคุมบนฝั่ง', r'การควบคุมจากระยะไกล', r'การตรวจสอบจากระยะไกล'
    ],
    'cybersecurity_resilience': [
        r'cybersecurity', r'cyber security', r'cyber[- ]attack', r'resilien',
        r'backup link', r'5G', r'LTE', r'communication redundancy',
        r'network resilience', r'encryption', r'authentication',
        r'keamanan siber', r'serangan siber', r'ketahanan jaringan',
        r'enkripsi', r'autentikasi',
        r'pag-atakeng cyber', r'katatagan ng network', r'enkripsyon',
        r'ความปลอดภัยไซเบอร์', r'การโจมตีไซเบอร์', r'การเข้ารหัส',
        r'ความทนทานของเครือข่าย'
    ],
    'regulation_procedures': [
        r'protocol', r'procedure', r'regulat', r'legal framework', r'guideline',
        r'standard operating procedure', r'NOTIFICATION',
        r'pilotage exemption', r'pilotage', r'authority approval',
        r'prosedur', r'aturan', r'kerangka hukum', r'pedoman',
        r'SOP', r'izin otoritas', r'pemanduan',
        r'proseso', r'regulasyon', r'balangkas na legal', r'patnubay',
        r'pag-apruba ng awtoridad',
        r'ข้อบังคับ', r'ขั้นตอน', r'กรอบทางกฎหมาย', r'แนวทางปฏิบัติ',
        r'การอนุมัติจากหน่วยงาน'
    ]
}
pillars = {k:[re.compile(p,flags=re.I) for p in v] for k,v in pillars.items()}

# ----------------------------
# Robust fetch with fallback
# ----------------------------
def extract_pdfminer(data):
    try: return pdfminer_extract(io.BytesIO(data))
    except: return ""

def extract_ocr(data, max_pages=3):
    try:
        images = convert_from_bytes(data, dpi=200, first_page=1, last_page=max_pages)
        return "\n".join(pytesseract.image_to_string(img) for img in images)
    except: return ""

def fetch_text(url, timeout=60):
    try:
        with requests.get(url, timeout=timeout, stream=True) as r:
            r.raise_for_status()
            ct = r.headers.get('Content-Type','')
            data = r.content
        if 'application/pdf' in ct or url.lower().endswith('.pdf'):
            text = ""
            with pdfplumber.open(io.BytesIO(data)) as pdf:
                for p in pdf.pages:
                    t = p.extract_text()
                    if t: text += "\n" + t
            if len(text.strip()) < 200:
                alt = extract_pdfminer(data)
                if len(alt.strip()) > len(text.strip()): text = alt
            if len(text.strip()) < 200:
                ocr = extract_ocr(data)
                if len(ocr.strip()) > len(text.strip()): text = ocr
        else:
            soup = BeautifulSoup(data, 'html.parser')
            for s in soup(['script','style','noscript']): s.decompose()
            text = soup.get_text(separator=' ', strip=True)
        return text
    except Exception as e:
        print(f"Fetch fail {url}: {e}")
        return ""

# ----------------------------
# Crawl all sources
# ----------------------------
years = range(2018,2025)
raw_texts = {}
evidence = {c:{y:{q:0 for q in pillars} for y in years} for c in sources}

print("Fetching...")
for c,urls in sources.items():
    for url in tqdm(urls, desc=c):
        txt = fetch_text(url)
        raw_texts[url] = txt
        found_years = set(re.findall(r'20(1[8-9]|2[0-4])', txt))
        found_years = {int('20'+y) for y in found_years} or set(years)
        for q,pats in pillars.items():
            matches = sum(1 for pat in pats if pat.search(txt))
            for y in found_years: evidence[c][y][q]+=matches
        if not txt or len(txt.strip())<50:
            print(f"[WARNING] {c} → kemungkinan teks minim: {url}")

# ----------------------------
# Evidence → ordinal scores
# ----------------------------
def to_score(c): return 0 if c==0 else 1 if c<=2 else 2
rows=[]
for c in sources:
    for y in years:
        row={'country':c,'year':y}
        for q in pillars: row[q]=to_score(evidence[c][y][q])
        rows.append(row)
df=pd.DataFrame(rows)

# ----------------------------
# Normalisasi hasil
# ----------------------------
cols=list(pillars.keys())
pillar_weights = {
    'technological_infrastructure': 1.2,
    'advanced_autonomous_technologies': 1.3,
    'interoperability_communication': 1.1,
    'onshore_operation_centers': 1.2,
    'cybersecurity_resilience': 1.0,
    'regulation_procedures': 0.7   # ← critical
}

df['score'] = sum(df[p] * w for p, w in pillar_weights.items())
df['SPRI_raw']=df['score']
df['SPRI_0_100']=MinMaxScaler((0,100)).fit_transform(df[['score']])
df['SPRI_0_100']=df['SPRI_0_100'].replace(0,0.1)

# ----------------------------
# Plot Line Chart per negara
# ----------------------------
plt.figure(figsize=(12,6))
markers=['o','s','D','^','v','P','X','*','h','+']
for i,c in enumerate(df['country'].unique()):
    sub=df[df['country']==c].sort_values('year')
    plt.plot(sub['year'],sub['SPRI_0_100'],
             marker=markers[i%len(markers)],
             label=c, linewidth=2, alpha=0.8)
plt.title("Smart Port Readiness Index (SPRI) ASEAN-5", fontsize=14)
plt.xlabel("Year");plt.ylabel("SPRI (0–100 scaled)")
plt.legend();plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout();plt.show()

# ----------------------------
# Extra: Normalized components in 2024 (line plot)
# ----------------------------
df2024 = df[df['year']==2024].set_index('country')
normalized = MinMaxScaler().fit_transform(df2024[cols])
df_norm = pd.DataFrame(normalized, columns=cols, index=df2024.index)

plt.figure(figsize=(12,6))
for c in df_norm.index:
    plt.plot(df_norm.columns, df_norm.loc[c], marker='o', label=c)
plt.title("Normalized components in 2024")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout();plt.show()

# ----------------------------
# Radar chart for latest year
# ----------------------------
def radar_chart(data, title="Radar Chart"):
    labels = list(data.columns)
    num_vars = len(labels)

    # Angles for each axis
    angles = np.linspace(0, 2*np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(7,7), subplot_kw=dict(polar=True))

    # Plot each country
    for idx in data.index:
        values = data.loc[idx].tolist()
        values += values[:1]   # close the loop
        ax.plot(angles, values, label=idx, linewidth=2)
        ax.fill(angles, values, alpha=0.15)

    # Axis labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=10)
    ax.set_yticks([0.25, 0.5, 0.75, 1.0])
    ax.set_yticklabels(["0.25","0.5","0.75","1.0"], fontsize=9)
    ax.set_ylim(0,1)

    # Title + legend
    ax.set_title(title, size=14, y=1.1)
    ax.legend(loc='upper left', bbox_to_anchor=(1.05,1.05))
    plt.tight_layout()
    plt.show()

# Prepare latest year normalized data
latest_year = df['year'].max()
df_latest = df[df['year'] == latest_year].set_index('country')
normalized = MinMaxScaler().fit_transform(df_latest[cols])
df_norm = pd.DataFrame(normalized, columns=cols, index=df_latest.index)

radar_chart(df_norm, f"Radar Chart of SPRI Components — {latest_year}")


# ----------------------------
# Save outputs
# ----------------------------
df.to_csv("spri_2018_2024.csv",index=False)
with open("spri_raw_texts.json","w") as f: json.dump(raw_texts,f)
print("Saved CSV + raw texts for manual validation")
```

```python
# ============================================================
# SMART PORT READINESS INDEX (SPRI) - REINFORCEMENT LEARNING SIMULATION
# Framework: Contextual Multi-Armed Bandit (LinUCB)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import random

# 1. CONFIGURATION & HYPERPARAMETERS
CSV_PATH = 'spri_2018_2024.csv'  # Ensure this matches your filename in Colab
RANDOM_SEED = 42
LINUCB_ALPHA = 1.0   # Exploration parameter
N_ROUNDS = 2000      # Total decision rounds
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

# 2. LOAD & PREPROCESS DATA
try:
    df = pd.read_csv(CSV_PATH)
    print(f"Successfully loaded {CSV_PATH}")
except FileNotFoundError:
    print(f"ERROR: {CSV_PATH} not found. Please upload the file to Colab.")
    raise

# Define the Action Space (The 6 SPRI Pillars)
arms = [
    'technological_infrastructure',
    'advanced_autonomous_technologies',
    'interoperability_communication',
    'onshore_operation_centers',
    'cybersecurity_resilience',
    'regulation_procedures'
]
K = len(arms)

# Calculate Reward Proxy (Next-year change in SPRI)
df = df.sort_values(['country', 'year']).reset_index(drop=True)
df['SPRI_next'] = df.groupby('country')['SPRI_0_100'].shift(-1)
df = df.dropna(subset=['SPRI_next']).copy()
df['delta'] = df['SPRI_next'] - df['SPRI_0_100']

# 3. CONTEXT FEATURE ENGINEERING (x_t)
# One-hot encoding for countries
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
country_ohe = ohe.fit_transform(df[['country']])

# Normalizing year and pillar scores
ctx_scaler = StandardScaler()
X_base = np.hstack([country_ohe, df[['year']].values, df[arms].values])
X_contexts = ctx_scaler.fit_transform(X_base)
context_dim = X_contexts.shape[1]

# 4. ENVIRONMENT SIMULATOR (Surrogate Reward Model)
# We train a model to "act" as the world, telling the agent the reward for its action
X_aug = np.repeat(X_contexts, K, axis=0)
actions_aug = np.tile(np.arange(K), len(X_contexts))
action_features = np.eye(K)[actions_aug]
X_train = np.hstack([X_aug, action_features])
y_train = np.repeat(df['delta'].values, K)

# Add synthetic signal: Investment in a pillar gives a 0.2 boost if pillar is already present
for i in range(len(X_contexts)):
    for a in range(K):
        y_train[i*K + a] += 0.2 * df[arms].iloc[i, a]

rf_env = RandomForestRegressor(n_estimators=100, max_depth=6, random_state=RANDOM_SEED)
rf_env.fit(X_train, y_train)

def get_env_reward(context_vec, arm_idx):
    a_onehot = np.zeros(K)
    a_onehot[arm_idx] = 1.0
    feat = np.concatenate([context_vec, a_onehot]).reshape(1, -1)
    return rf_env.predict(feat)[0] + np.random.normal(0, 0.05)

# 5. THE RL AGENT (LinUCB Algorithm)
class LinUCBAgent:
    def __init__(self, d, K, alpha):
        self.A = [np.eye(d) for _ in range(K)]
        self.b = [np.zeros((d, 1)) for _ in range(K)]
        self.alpha = alpha
        self.K = K

    def select_arm(self, x_t):
        x_t = x_t.reshape(-1, 1)
        p = np.zeros(self.K)
        for a in range(self.K):
            A_inv = np.linalg.inv(self.A[a])
            theta = A_inv @ self.b[a]
            uncertainty = self.alpha * np.sqrt(x_t.T @ A_inv @ x_t)
            p[a] = theta.T @ x_t + uncertainty
        return np.argmax(p)

    def update(self, arm, x_t, reward):
        x_t = x_t.reshape(-1, 1)
        self.A[arm] += x_t @ x_t.T
        self.b[arm] += reward * x_t

# 6. EXECUTE SIMULATION
agent = LinUCBAgent(context_dim, K, LINUCB_ALPHA)
hist_lincb, hist_random, hist_oracle = [], [], []
cum_lincb, cum_random, cum_oracle = 0, 0, 0
counts = np.zeros(K)

for t in range(N_ROUNDS):
    idx = np.random.randint(0, len(X_contexts))
    x_t = X_contexts[idx]

    # LinUCB Policy
    arm_l = agent.select_arm(x_t)
    reward_l = get_env_reward(x_t, arm_l)
    agent.update(arm_l, x_t, reward_l)
    cum_lincb += reward_l
    counts[arm_l] += 1

    # Random Policy (Baseline)
    reward_r = get_env_reward(x_t, np.random.randint(0, K))
    cum_random += reward_r

    # Oracle Policy (Perfect Knowledge)
    all_rewards = [get_env_reward(x_t, a) for a in range(K)]
    cum_oracle += max(all_rewards)

    hist_lincb.append(cum_lincb)
    hist_random.append(cum_random)
    hist_oracle.append(cum_oracle)

# 7. RESULTS & VISUALIZATION
plt.figure(figsize=(10, 5))
plt.plot(hist_lincb, label=f"LinUCB (Alpha={LINUCB_ALPHA})", linewidth=2)
plt.plot(hist_random, label="Random Policy", linestyle='--')
plt.plot(hist_oracle, label="Oracle Policy", linestyle=':')
plt.title("SPRI Predictive Utility: Cumulative Reward Comparison")
plt.xlabel("Simulation Round")
plt.ylabel("Cumulative Reward (Δ SPRI)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("\n--- Simulation Summary ---")
for i, arm in enumerate(arms):
    print(f"Pillar: {arm:35} | Times Chosen: {int(counts[i])}")
```
